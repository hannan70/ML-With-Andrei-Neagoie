```python
car_sales = pd.read_csv("car-sales.csv")

car_sales
```

```python
car_sales.rename(columns={"Make": "make", "Colour": "color", "Odometer (KM)": "odometer", "Doors": "door", "Price":"price"}, inplace=True)
```

```python
car_sales
```

```python
car_sales['price'] = car_sales['price'].str.replace(r'[\$\,\.]', '', regex=True)
```

```python
car_sales['price'] = car_sales['price'].str[:-2]
```

```python
car_sales['price']  = car_sales['price'].astype(int)
```

```python
car_sales
```

```python
car_sales['price'].dtype
```

```python
car_sales['Sale Date'] = pd.date_range('1/1/2024', periods=len(car_sales))
```

```python
car_sales
```

```python
x = car_sales['price']
y = car_sales['Sale Date']

fig, ax = plt.subplots()
ax.scatter(x, y)
```

```python
car_sales['Total Sale'] = car_sales['price'].cumsum()
car_sales
```

```python
car_sales.plot(x="Sale Date", y="Total Sale")
```

```python
car_sales.plot(x="odometer", y="price", kind="scatter")
```

```python
car_sales.plot(x="make", y="price", kind="bar")
```

```python
heart_disease = pd.read_csv("heart-disease.csv")
heart_disease
```

```python
over_50 = heart_disease[ heart_disease['age'] > 50]

over_50
```

```python
# pyplot method
over_50.plot(x="age", y="chol", kind="scatter", c="target");
```

```python
# oop method
fig, ax = plt.subplots(figsize=(10, 6))

# Plot the data
scatter = ax.scatter(x=over_50['age'],
                    y=over_50['chol'],
                    c=over_50['target'])

# Customize  the plot
ax.set(title="Heart Disease and Cholesterol Levels",
      xlabel="Age",
      ylabel="Cholesterol")

# Add a Legend
ax.legend(*scatter.legend_elements(), title="target")

# Add horizontal line
ax.axhline(over_50['chol'].mean(),
          linestyle ="--",
          color="r");

```

```python
plt.style.use("seaborn-v0_8-whitegrid")
# subplot of chol, age and thalach
fig, (ax0, ax1) = plt.subplots(nrows=2,
                              ncols=1,
                              figsize=(10, 10),
                              sharex=True)

# Add data to ax0
scatter = ax0.scatter(x=over_50['age'],
                     y=over_50['chol'],
                     c=over_50['target'],
                     cmap="winter") # change the color theme

# Customize ax0
ax0.set(title="Heart Disease and Cholesterol Levels",
       ylabel="Cholesterol")
# Change the x axis limits
ax0.set_xlim([50, 80])

# Add Legend
ax0.legend(*scatter.legend_elements(), title="Target")

# Add a mean line
ax0.axhline(over_50['chol'].mean(),
          linestyle="--",
          color="r")

# Add a data to ax1
scatter = ax1.scatter(x=over_50['age'],
                     y=over_50['thalach'],
                     c=over_50['target'],
                     cmap="winter")

# Customize to ax1
ax1.set(title="Heart Disease and Max Heart Rate",
       xlabel="Age",
       ylabel="Max Heart Rate")

# Change the x axis limits
ax1.set_xlim([50, 80])
ax1.set_ylim([60, 200])

# Add Legend 
ax1.legend(*scatter.legend_elements(), title="Target")

# add minline
ax1.axhline(over_50['thalach'].mean(),
           linestyle="--",
           color="r")

# Add suptitle
fig.suptitle("Heart Disease Analysis", fontsize=20, fontweight="bold");

```
